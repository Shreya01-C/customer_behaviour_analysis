# Treating the null values

## size

```python
df['Size'].unique()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Item Purchased')['Size'].unique()
```

```python
# Treating null values or replacing all the values in Electronics category
```

```python
df.loc[df['Category'] == 'Electronics', 'Size'] = "Not Applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all the values in Accessories category
df.loc[df['Category'] == 'Accessories', 'Size'] = "Free Size"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all the values in Accessories category
df.loc[df['Category'] == "Footwear", 'Size'] = "Not Available"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all the values in Clothing category
df.loc[(df['Category'] == "Clothing") & (df["Size"].isnull()), "Size"] = df[df["Category"] == "Clothing"]["Size"].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Filling previous purchases wih zero(0)
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```
